# Laser pulse propagation in optical fiber (GVD only case)

We are asked to **analytically solve** the **fractional partial differential equation**:

$$
\frac{\partial^\alpha A(z,t)}{\partial z^\alpha} = i \frac{\beta_2}{2} \frac{\partial^2 A(z,t)}{\partial t^2}, \quad 0 < \alpha < 1,
$$

where the **Caputo fractional derivative** is taken with respect to **spatial variable** $z$, and the initial condition is a **chirped Gaussian pulse**:

$$
A(0, t) = A_0(t) = A_p \exp\left(-\frac{t^2}{2T_0^2} + i C t^2\right).
$$

---

### ðŸ” Step-by-step plan:

1. **Take Fourier transform** in time variable $t$ to simplify the RHS (since time derivatives turn into multiplication in Fourier domain).
2. Solve the resulting **fractional ODE** in $z$ for each frequency component using known results for fractional Caputo derivatives.
3. **Invert the Fourier transform** to get back $A(z, t)$.

---

### Step 1: Fourier transform in time

Define the **temporal Fourier transform** as:

$$
\hat{A}(z, \omega) = \mathcal{F}_t[A(z, t)] = \int_{-\infty}^{\infty} A(z, t) e^{i \omega t} \, dt.
$$

Applying it to both sides of the equation:

$$
\mathcal{F}_t\left[\frac{\partial^\alpha A}{\partial z^\alpha}\right] = i \frac{\beta_2}{2} \mathcal{F}_t\left[\frac{\partial^2 A}{\partial t^2}\right],
$$

But the Fourier transform of $\frac{\partial^2 A}{\partial t^2}$ is $-\omega^2 \hat{A}(z, \omega)$, so:

$$
\frac{\partial^\alpha \hat{A}(z, \omega)}{\partial z^\alpha} = -i \frac{\beta_2}{2} \omega^2 \hat{A}(z, \omega).
$$

This is a **fractional ODE** in $z$, for each fixed $\omega$:

$$
\frac{d^\alpha \hat{A}(z, \omega)}{dz^\alpha} = -i \frac{\beta_2}{2} \omega^2 \hat{A}(z, \omega), \quad \hat{A}(0, \omega) = \hat{A}_0(\omega).
$$

---

### Step 2: Solve the fractional ODE

This is a **linear fractional differential equation** with constant coefficients.

The Caputo fractional ODE:

$$
\frac{d^\alpha y(z)}{dz^\alpha} = \lambda y(z), \quad y(0) = y_0,
$$

has the solution:

$$
y(z) = y_0 E_\alpha(\lambda z^\alpha),
$$

where $E_\alpha(\cdot)$ is the **Mittag-Leffler function**:

$$
E_\alpha(x) = \sum_{k=0}^\infty \frac{x^k}{\Gamma(\alpha k + 1)}.
$$

Thus, our solution is:

$$
\hat{A}(z, \omega) = \hat{A}_0(\omega) \cdot E_\alpha\left(-i \frac{\beta_2}{2} \omega^2 z^\alpha \right).
$$

---

### Step 3: Compute the Fourier transform of the initial condition

Recall:

$$
A_0(t) = A_p \exp\left( -\frac{t^2}{2 T_0^2} + i C t^2 \right) = A_p \exp\left( -\frac{1 - 2iCT_0^2}{2T_0^2} t^2 \right).
$$

This is a **chirped Gaussian**, and its Fourier transform is well known:

$$
\hat{A}_0(\omega) = A_p T_0 \sqrt{2\pi} \cdot \frac{1}{\sqrt{1 - 2i C T_0^2}} \exp\left( -\frac{T_0^2 \omega^2}{2 (1 - 2i C T_0^2)} \right).
$$

---

### Step 4: Total solution in frequency domain

$$
\hat{A}(z, \omega) = \hat{A}_0(\omega) \cdot E_\alpha\left(-i \frac{\beta_2}{2} \omega^2 z^\alpha \right).
$$

Now take the **inverse Fourier transform** to recover $A(z, t)$:

$$
A(z, t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} \hat{A}_0(\omega) \cdot E_\alpha\left(-i \frac{\beta_2}{2} \omega^2 z^\alpha \right) e^{-i \omega t} d\omega.
$$

This integral defines the **analytical solution**.

---

### âœ… Final Answer

The **analytical solution** is:

$$
A(z, t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} \hat{A}_0(\omega) \cdot E_\alpha\left(-i \frac{\beta_2}{2} \omega^2 z^\alpha \right) e^{-i \omega t} d\omega,
$$

with

$$
\hat{A}_0(\omega) = A_p T_0 \sqrt{2\pi} \cdot \frac{1}{\sqrt{1 - 2i C T_0^2}} \exp\left( -\frac{T_0^2 \omega^2}{2 (1 - 2i C T_0^2)} \right),
$$

and $E_\alpha(\cdot)$ is the **Mittag-Leffler function**.
