# Laser pulse propagation in optical fiber (attenuation only case)

To solve the fractional equation

$$
\frac{\partial^\alpha A(z,t)}{\partial z^\alpha} = -\alpha_{att} A(z,t), \quad 0 < \alpha < 1,
$$

with the **Caputo fractional derivative** in the spatial variable $z$, and initial condition at $z = 0$:

$$
A(0, t) = A_0(t) = A_p \exp\left(-\frac{t^2}{2T_0^2} + i C t^2\right),
$$

we proceed **analytically** using the **Laplace transform method** in the spatial variable $z$, exploiting known properties of the Caputo derivative.

---

### ðŸ§® Step 1: Caputo Fractional Derivative and Laplace Transform

The Caputo fractional derivative of a function $A(z,t)$ of order $\alpha \in (0,1)$ with respect to $z$ is defined as:

$$
\frac{\partial^\alpha A(z,t)}{\partial z^\alpha} = \frac{1}{\Gamma(1 - \alpha)} \int_0^z \frac{\partial A(\xi,t)}{\partial \xi} \frac{d\xi}{(z - \xi)^\alpha}
$$

The Laplace transform in $z$, with $\mathcal{L}_z \{A(z, t)\} = \tilde{A}(s, t)$, gives:

$$
\mathcal{L}_z \left\{ \frac{\partial^\alpha A(z, t)}{\partial z^\alpha} \right\} = s^\alpha \tilde{A}(s,t) - s^{\alpha - 1} A(0,t)
$$

Then, applying this to the fractional equation:

$$
s^\alpha \tilde{A}(s,t) - s^{\alpha - 1} A_0(t) = -\alpha_{att} \tilde{A}(s,t)
$$

---

### ðŸ§© Step 2: Solve for $\tilde{A}(s,t)$

$$
\tilde{A}(s,t) \left(s^\alpha + \alpha_{att} \right) = s^{\alpha - 1} A_0(t)
$$

$$
\tilde{A}(s,t) = \frac{s^{\alpha - 1}}{s^\alpha + \alpha_{att}} A_0(t)
$$

---

### ðŸ“¤ Step 3: Inverse Laplace Transform

The Laplace inverse of $\frac{s^{\alpha - 1}}{s^\alpha + \alpha_{att}}$ is known:

$$
\mathcal{L}^{-1} \left\{ \frac{s^{\alpha - 1}}{s^\alpha + \alpha_{att}} \right\} = E_\alpha(-\alpha_{att} z^\alpha)
$$

where $E_\alpha(\cdot)$ is the **Mittag-Leffler function**:

$$
E_\alpha(-\alpha_{att} z^\alpha) = \sum_{k=0}^\infty \frac{(-\alpha_{att} z^\alpha)^k}{\Gamma(\alpha k + 1)}
$$

Thus:

$$
A(z, t) = A_0(t) \cdot E_\alpha(-\alpha_{att} z^\alpha)
$$

---

### âœ… Final Analytical Solution

$$
\boxed{
A(z, t) = A_p \exp\left(-\frac{t^2}{2 T_0^2} + i C t^2\right) \cdot E_\alpha(-\alpha_{att} z^\alpha)
}
$$

This is the **analytic solution** in terms of the **Mittag-Leffler function**, which naturally generalizes the exponential decay in the standard ($\alpha = 1$) case.

---

### ðŸ“Œ Notes:

* When $\alpha = 1$, $E_1(-\alpha_{att} z) = \exp(-\alpha_{att} z)$, and the solution reduces to the usual exponential attenuation.
* For $0 < \alpha < 1$, the Mittag-Leffler function represents **slower-than-exponential** decay, modeling anomalous attenuation or sub-diffusive behaviors.
